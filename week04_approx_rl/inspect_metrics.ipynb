```python
import torch

dueling = torch.load('logs/dueling_metrics.pth')
dqn = torch.load('logs/dqn_metrics.pth')
```

```python
import numpy as np
import utils
import matplotlib.pyplot as plt
%matplotlib inline


def plot_history(mean_rw_history, td_loss_history, 
                 initial_state_v_history, grad_norm_history):
    plt.figure(figsize=[16, 9])

    plt.subplot(2, 2, 1)
    plt.title("Mean reward per life")
    plt.plot(mean_rw_history)
    plt.grid()

    assert not np.isnan(td_loss_history[-1])
    plt.subplot(2, 2, 2)
    plt.title("TD loss history (smoothened)")
    plt.plot(utils.smoothen(td_loss_history))
    plt.grid()

    plt.subplot(2, 2, 3)
    plt.title("Initial state V")
    plt.plot(initial_state_v_history)
    plt.grid()

    plt.subplot(2, 2, 4)
    plt.title("Grad norm history (smoothened)")
    plt.plot(utils.smoothen(grad_norm_history))
    plt.grid()

    plt.show()
    return
    
    
```

```python
plot_history(dqn['mean_rw_history'], dqn['td_loss_history'],
            dqn['grad_norm_history'], dqn['initial_state_v_history'])
```

```python

plot_history(dueling['mean_rw_history'], dueling['td_loss_history'],
            dueling['grad_norm_history'], dueling['initial_state_v_history'])
```
